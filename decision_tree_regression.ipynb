# Decision tree regression:

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values  #Usially done for independent variaable X.
y = dataset.iloc[:, 2].values
```

```python
# Splitting the dataset into the Training set and Test set
"""from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)"""
```

```python
# Feature Scaling
"""from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train)"""
```

```python
# Fitting the Regression Model(decision tree regression) to the dataset
from sklearn.tree import DecisionTreeRegressor
d = DecisionTreeRegressor()
d.fit(X,y)
```

```python
#Predicting the result on 6.5 years:
y_pred = d.predict(np.array([[6.5]]))
y_pred
```

```python
# Visualising the Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, d.predict(X), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regressor)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Regression results (for higher resolution and smoother curve)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, d.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regresso at 0.1)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
